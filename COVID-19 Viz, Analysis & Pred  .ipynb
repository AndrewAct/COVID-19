## The newest data

```python
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv',parse_dates=['Last Update'])
df.rename(columns={'ObservationDate':'Date', 'Country/Region':'Country'}, inplace=True)

print(df.head())
```

```python
confirmed = df['Confirmed']
deaths = df['Deaths']
recovered = df['Recovered']
```

```python
date = df['Date']
```

## View the cases in China

```python
df.query('Country =="Mainland China"').groupby("Last Update")[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

## View the cases by countries 

```python
df.groupby("Country")[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

## Visualization

```python
df.groupby('Date').sum()
```

```python
confirmed = df.groupby('Date').sum()['Confirmed'].reset_index()
deaths = df.groupby('Date').sum()['Deaths'].reset_index()
recovered = df.groupby('Date').sum()['Recovered'].reset_index()
```

```python
fig, axs = plt.subplots(3)
fig.suptitle('Confimred, Deaths and Recovered Resprect to Timw')
axs[0].plot(confirmed['Date'], confirmed['Confirmed'])
axs[1].plot(deaths['Date'], deaths['Deaths'])
axs[2].plot(recovered['Date'], recovered['Recovered'])
```

```python
fig = plt.figure()
plt.plot(confirmed['Date'], confirmed['Confirmed'], '--')
plt.plot(deaths['Date'], deaths['Deaths'], '-')
plt.plot(recovered['Date'], recovered['Recovered'], '.')
plt.title('Confimred, Deaths and Recovered Resprect to Timw')
plt.xlabel("Date")
plt.ylabel("Number of people")
```
